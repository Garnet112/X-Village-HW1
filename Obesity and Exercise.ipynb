2008~2013年來18歲以上過重及肥胖率:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



url = 'http://data.hpa.gov.tw/dataset/c335205a-15ab-4679-a97e-12d0dec4cf3e/resource/a8fd7d74-611f-4838-9cf6-3c1784b93fc9/download/des0791.4.1441818.csv'
data = pd.read_csv(url, encoding='utf-8')
print(data.head())

plt.figure(figsize=(15, 5))
x = data['依年度別分by year']
y = data['男Male (%)'] 
plt.plot(x, y, '-o',mfc='orange')
plt.title("男Male")
plt.show()
```

以上圖表為歷年來18歲以上過重男子所占百分比，從圖表看來我們可以很明顯地知道肥胖人口有逐漸上升的趨勢

```python
plt.figure(figsize=(15, 5))
x = data['依年度別分by year'] 
y = data['女Female (%)'] 
plt.plot(x, y, '-o',mfc='red')
plt.title("Female")
plt.show()
```

以上圖表為歷年來18歲以上過重女子所占百分比，從2009年之後也是逐漸上升的趨勢

```python
url = 'http://data.hpa.gov.tw/dataset/363bc2a5-0930-40bc-89ab-b35cf1d49095/resource/bfc861a3-71b3-4d47-8f73-4a03d86c5e78/download/des0791.4.8494.csv'
data = pd.read_csv(url, encoding='utf-8')
print(data.head())

plt.figure(figsize=(15, 5))
height = data['40公斤運動30分鐘所消耗的熱量 (大卡)'] 
bars = range(0,data.shape[0]) 
y_pos = np.arange(len(bars))
 
# Create bars
plt.bar(y_pos, height)
 
# Create names on the x-axis
plt.xticks(y_pos, bars)
 
plt.show()
```

從以上圖表我們可以知道每小時快跑16公里是能最快消耗熱量的運動方式
